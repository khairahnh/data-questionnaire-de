## c. Show the list of customers who never purchased anything


```python
customers_invoices = customers.join(invoices.set_index('customer_id'), on='id',
               lsuffix='_customers', rsuffix='_invoices')

customers_invoices




#Perform left join for both dataframes 
# join two dataframes invoices and customers
# set the index to `customer_id` from the customers dataframe and using `id` column as the common key
# the lsuffix and rsuffix parameters specify suffixes to add to overlapping column names from left(customers) and right(invoices) dataframes
```

```python
#Filter customers_invoices (joined dataframes)
#rows where the corresponding value in the id_invoices column is missing (NaN) will be included

customers_never_purchased= customers_invoices[customers_invoices['id_invoices'].isna()]



#select `name` and `total` column from customers_invoices dataframe
customers_never_purchased[['name','total']]
```

# d. list of book purchased with the users.


```python
#perform left join between customers_invoices and invoice_lines

customers_invoice_lines = customers_invoices.join(invoice_lines.set_index('invoice_id'), 
                                                 on='id_invoices',
                                                 lsuffix='_customers_invoices', 
                                                 rsuffix='_invoice_lines')
customers_invoice_lines



# Perform left join from existing joined dataframe `customers_invoices` with `invoice_lines` 
# set the index to `invoice_id` from the `invoice_lines` dataframe and using `id_invoices`column from `customers_invoices` as the common key
# the lsuffix and rsuffix parameters specify suffixes to add to overlapping column names from left(customers_invoices) and right(invoice_lines) dataframes
```

```python
#filter customers who have purchased by filtering description not null
book_purchased_customers = customers_invoice_lines[customers_invoice_lines['description'].notnull()]

book_purchased_customers[['name', 'description']]
```
